```python
#create data cleaning function
def clean_movie(movie):
    #create a non-diestructive copy
    movie=dict(movie)
    #make empty dict to hold the alt titles
    alt_titles={}
    #loop through list of alt. title keys
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        #check if current key exists in movie object
        if key in movie:
            #if so, remove the key-value pair and add to alt titles
            alt_titles[key]=movie[key]
            movie.pop(key)
    #add alt titles dict to movie object        
    if len(alt_titles)>0:
        movie['alt_titles']=alt_titles
            
    return movie
```

```python
#create list of cleaned movies
clean_movies=[clean_movie(movie) for movie in wiki_movies]
```

```python
wiki_movied_df=pd.DataFrame(clean_movies)
sorted(wiki_movies_df.columns.tolist())
```
